## Teilaufgabe 1: Wie viele Ausleihen gab es pro Bibliothek?

```python
df['Total Checkouts'].sum()
```

```python
checkouts_library = df[['Home Library Definition','Total Checkouts']]
checkouts_library.sort_values('Home Library Definition')
```

```python
#Ausleihen pro Bibliothek Total

checkouts_library.groupby(by='Home Library Definition').sum()
```

## Teilaufgabe 2: Wie viele Teenager haben sich im Jahr 2009 in der Main-Bibliothek eingeschrieben? 

```python
teen=df.loc[df['Patron Type Definition'] == 'Teen',:]
teen
```

```python
teen_main = teen.loc[teen['Home Library Definition'] == 'Main',:]
teen_main
```

```python
teen_main_2009 = teen_main.loc[teen_main['Year Patron Registered'] == 2009]
teen_main_2009
```

```python
teen_main_2009.mean()
```

```python
sns.relplot(x='Total Checkouts', y='Total Renewals', data=teen_main_2009,
            height=5, aspect=3)
```
