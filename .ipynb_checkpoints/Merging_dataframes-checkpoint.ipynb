```python
#sobald es um Tabellen geht --> PANDAS!!

import pandas as pd
```

```python
doi_df_journal = pd.DataFrame({
    "DOI": [
        "10.1371/journal.pcbi.1004668",
        "10.21105/joss.01035",
        "10.12688/f1000research.18866.2",
        "10.1038/s41598-019-52881-4",
        "10.1186/s12859-019-3171-0"],
    "journal": [
        "PLoS Comput Biol",
        "JOSS",
        "F1000Res",
        "Sci Rep",
        "BMC Bioinformatics"]})

doi_df_first_author = pd.DataFrame({
    "DOI": [
        "10.1371/journal.pcbi.1004668",
        "10.21105/joss.01035",
        "10.21105/joss.01006",
        "10.12688/f1000research.18866.2",
        "10.1186/s12859-019-3171-0"],
    "first_author": [
        "Blischak",
        "Sparks",
        "Granger",
        "Thang",
        "Chen"]})

```

```python
doi_df_journal
```

```python
doi_df_first_author
```

```python
#inner join (default) nimmt jene, die überlappen, alles was nicht passt, fällt raus. 

doi_df_first_author.merge(doi_df_journal, on="DOI")

#wenn die Spaltennamen nicht gleich sind kann man left_on und right_on verwenden
```

```python
#left: ich will alles, was auf der linken Seite ist, muss da sein. Wo rechts nicht steht, wird NaN eingetragen. 

doi_df_first_author.merge(doi_df_journal, on="DOI", how="left")
```

```python
#right: ich will alles, was auf der rechten Seite ist, muss da sein. Wo links nicht steht, wird NaN eingetragen. 

doi_df_first_author.merge(doi_df_journal, on="DOI", how="right")
```

```python
#outer join. nimmt alles rein und wo nichts steht kommt NaN

doi_df_first_author.merge(doi_df_journal, on="DOI", how="outer")
```
