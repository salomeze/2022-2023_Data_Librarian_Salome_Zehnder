```python
small_apcs = apcs[["euro", "publisher"]]
#small_apcs ist jetzt nur ein Teil von der Hauptliste, nämlich nur mit den Spalten euro und publisher. Wir kriegen ein neuen Dataframe der kleiner ist als das Original
```

```python
def mod_publisher(publisher):
    publisher = publisher.replace(" ", "_")
    publisher = publisher.upper()
    return publisher
```

```python
#neue Spalte und damit soll etwas gemacht werden mit apply, apply brauch eals Argument eine Funktion.


small_apcs["publisher_modified"] = small_apcs["publisher"].apply(mod_publisher)
```
