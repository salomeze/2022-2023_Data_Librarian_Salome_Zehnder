## Frage 1: Top10 Bestsellerliste der Prints

Welche 10 Print-Titel wurden am meisten ausgeliehen? 


```python
#zuerst mache ich es mir übersichtlicher
```

```python
checkouts_df = pd.DataFrame.reindex(checkouts_df,columns = ['title', 'creator', 'checkouts','materialtype'])
```

```python
checkouts_df
```

```python
checkouts_books_df = checkouts_df[checkouts_df["materialtype"] == "BOOK"]
```

```python
checkouts_books_df.shape
```

```python
checkouts_books_df.to_csv("checkout_books_only.csv")
```

```python
checkouts_books_df.sort_values(by='checkouts', ascending=False)
```

```python
#Für die Top10 speichere ich die sortierte Liste neu ab und zeige mir die ersten 10 Zeilen an. 
```

```python
checkouts_sorted_books_df = checkouts_books_df.sort_values(by='checkouts', ascending=False)
```

```python
checkouts_sorted_books_df.to_csv("checkout_sorted_books_only.csv")
```

```python
checkouts_sorted_books_df.head(10)
```

```python
#ich speichere die Top 10 ebenfalls neu ab, um damit eine schöne Grafik zu erstellen
#die Grafik könnte ich wohl auch ohne erneutes Zwischenspeichern anzeigen, es erscheint mir auf meiner Wissensbasis aber einfacher
```

```python
top10_df = checkouts_sorted_books_df.iloc[0:10]
```

```python
top10_df.to_csv("top10_df.csv")
```

```python
import matplotlib.pyplot as plt
```

```python
top10_df.plot(x="title", y="checkouts", kind="bar")

plt.ylabel("checkouts", 
           family='serif', 
           color='b', 
           weight='normal', 
           size = 15)
plt.xlabel("title", 
           family='serif', 
           color='r', 
           weight='normal', 
           size = 15)
```
