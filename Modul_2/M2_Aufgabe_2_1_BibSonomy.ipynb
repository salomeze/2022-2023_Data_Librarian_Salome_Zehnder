```python


# einlesen von Daten aus Bibsonomy
import requests
url = "https://www.bibsonomy.org/json/search/Bibliothek?items=1000&duplicates=merged"
result = requests.get(url) # result ist ein requests.models.Response Objekt
data = result.json() # das Response-Objekt bietet diese nützliche Methode an

```

```python
data.keys()
```

```python
type(data["items"])
```

```python
data["items"][5]
```

```python
df_bibsonomy = pd.DataFrame(data["items"])
```

```python
df_bibsonomy
```

```python
df_bibsonomy["type"].value_counts()
```

```python
df_bibsonomy.head(1)
```

```python
df_bibsonomy.columns
```

```python
df_bibsonomy=pd.DataFrame.reindex(df_bibsonomy,columns = ['type', 'label','description', 'pub-type','booktitle','journal','year','author','authors','publisher','isbn','issn','language','doi'])
```

```python
df_bibsonomy.columns
```

```python
df_bibsonomy
```

```python
df_bibsonomy_pubs = df_bibsonomy[df_bibsonomy["type"] == "Publication"]
```

```python
df_bibsonomy_pubs.shape
```

```python
df_bibsonomy_pubs.to_json("bibsonomy_pubs.json")
```

```python
df_bibsonomy_pubs.to_csv("bibsonomy_pubs.csv")
```
